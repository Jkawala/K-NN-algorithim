# Knn Algorithm

```python
from sklearn import neighbors
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import model_selection
import pandas as pd
import numpy as np

def question1Part1(train_features, train_labels, test_features, test_labels, dmetric, w):
    
    # Explore and plot results for combinations of hyper-parameters (k and d metric)
    allResults = []
    for kValue in range(1, 100):

        knn = neighbors.KNeighborsClassifier(n_neighbors=kValue, metric= dmetric, weights = w)
        knn = knn.fit(train_features, train_labels)
        accuracy = knn.score(test_features, test_labels)
        allResults.append(accuracy)
        #print (kValue, accuracy)

    plt.plot(allResults)
    plt.show()
```

### Uniform Weight and Euclidean distance 

```python
# Load dataset into a NumPy array
cancerTrain = np.genfromtxt("cancerTrain.csv", delimiter=",")
cancerTest = np.genfromtxt("cancerTest.csv", delimiter=",")
    
# Extract features and label data
train_features = cancerTrain[:, :-1]
train_labels = cancerTrain[:, -1]

test_features = cancerTest[:, :-1]
test_labels = cancerTest[:, -1]
    
#‘uniform’ : uniform weights. All points in each neighborhood are weighted equally (default).
# Evaluate the performance of kNN
question1Part1(train_features, train_labels, test_features, test_labels, "euclidean", "uniform")
# Evaluate the performance of kNN
```

### Uniform Weight and Manhattan distance 

```python
#‘uniform’ : uniform weights. All points in each neighborhood are weighted equally (default).
# Evaluate the performance of kNN
question1Part1(train_features, train_labels, test_features, test_labels, "manhattan", "uniform")
```
